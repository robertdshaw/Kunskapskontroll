## Query 1: Personal Resurser

##### 1a. AW verksamhetsområden

```python
# Här bestämde jag mig för att ansluta distinkta anställda för att kunna se både företagets divisionsindelning som grupper, 
# och använda distinkt för att undvika eventuella dubbletter. Jag använde union för att snabbt jämföra storleken på varje 
# division med den totala storleken på företaget, inom ett enda dataset.

# Lista av verksamhetsområden i AdventureWorks
query_1a_områden = """
SELECT	D.GroupName AS [AdventureWorks Divisions]
    , COUNT(DISTINCT E.BusinessEntityID) AS [No. of Employees]
FROM HumanResources.Department AS D
	LEFT JOIN 
    HumanResources.EmployeeDepartmentHistory AS EDH 
	ON D.DepartmentID = EDH.DepartmentID
	LEFT JOIN 
    HumanResources.Employee AS E 
	ON EDH.BusinessEntityID = E.BusinessEntityID
GROUP BY 
    D.GroupName

UNION 

SELECT 'Total' AS [AdventureWorks Divisions]
    , (SELECT COUNT(DISTINCT BusinessEntityID) FROM HumanResources.Employee) AS [No. of Employees]
FROM  HumanResources.Department;
"""  
df1a=pd.read_sql(query_1a_områden, con=connection)
df1a
```

##### 1b. AW Anställda

```python
# Jag bestämde mig för att kontrollera om företaget har fokus på jämställdhet i sin rekryteringsprocess. 
# Jag använde CAST för att konvertera antalet anställda från ett heltal till ett flyttal eftersom jag ville visualisera 
# uppdelningen som en procentandel i ett cirkeldiagram. 

query_1b_anstallda = """
SELECT Gender
    , COUNT(*) AS Employees
    , ROUND(CAST(COUNT(*) AS FLOAT) / (SELECT CAST(COUNT(*) AS FLOAT) FROM HumanResources.Employee) * 100, 0) AS PercentageOfTotal
FROM HumanResources.Employee
GROUP BY Gender;
"""
df1b = pd.read_sql(query_1b_anstallda, con=connection)
df1b
```

```python
# Skapa ett cirkeldiagram
plt.figure(figsize=(8, 8))
plt.pie(df1b['PercentageOfTotal'], labels=df1b['Gender'], autopct='%1.1f%%', startangle=140)
plt.title('Employee Distribution by Gender')
plt.show()
```

## Query 2: Kunder

```python
# Jag körde denna kundkod för att visa hur stort företaget är som en bra översikt, och det är tydligt att det är 
# ett ganska stort företag med nära 20 000 kunder

query_2a_kunder = """

SELECT COUNT(DISTINCT CustomerID) AS TotalKunder
FROM Sales.Customer;
""" 
df_2a = pd.read_sql(query_2a_kunder, con=connection)
df_2a
```

```python
# Jag organiserade kunderna med en serie sammanfogningar för att pussla ihop en grupp där kunderna kommer ifrån geografiskt.  
# Så denna query kopplar samman flera tabeller för att länka varje kund till deras geografiska plats. Det är ett internationellt
# företag.

# DISTINCT innebär att varje kund endast räknas en gång, vilket ger ett korrekt antal unika kunder. 
# I annat fall skulle COUNT()-funktionen räkna varje förekomst av ett CustomerID, vilket skulle kunna leda till ett överdrivet 
# antal om en kund förekommer mer än en gång i det sammanfogade datasetet.

# Den går vidare med tabellerna BusinessEntityAddress, Address, StateProvince och CountryRegion för att få landspositionen för 
# varje kund.# Queryn räknar unika kunder (CustomerID) i varje land och grupperar resultaten därefter.
# Resultaten sorteras efter antalet kunder i varje land, vilket visar var majoriteten av kundbasen finns.

query_2b_kunder = """
SELECT A.Name AS Country   
    , COUNT(DISTINCT C.CustomerID) AS NumberOfCustomers -- 
FROM Sales.Customer AS C
LEFT JOIN Person.BusinessEntityAddress AS BEA ON C.CustomerID = BEA.BusinessEntityID
LEFT JOIN Person.Address AS AD ON BEA.AddressID = AD.AddressID
LEFT JOIN Person.StateProvince AS SP ON AD.StateProvinceID = SP.StateProvinceID
LEFT JOIN Person.CountryRegion AS A ON SP.CountryRegionCode = A.CountryRegionCode 
GROUP BY A.Name 
ORDER BY NumberOfCustomers DESC; 
""" 
df_2b = pd.read_sql(query_2b_kunder, con=connection)
df_2b
```

* Det finns 9547 utan adress och dessa är sannolikt antingen offline-försäljning där adress inte är en obligatorisk kategori.

```python
# Med hjälp av Dataedo AdventureWorks Data Dictionary fick jag följande information:
# SC = Butikskontakt, 
# IN = Enskild kund
# SP = Säljare
# EM = Anställd (icke-säljare)
# VC = Leverantörskontakt
# GC = Allmän kontakt

# Så jag antog att bortsett från anställda (SP och EM), så är resten en uppdelning mellan B2B- eller B2C-kunder.

# Denna query använder en CASE WHEN-sats för att sortera mellan personkategorier som B2B eller B2C och exkluderar anställda 

query_2c_kunder = """
SELECT Customers.[People Categories], SUM(Customers.People) AS People
FROM (
    SELECT CASE PersonType
        WHEN 'IN' THEN 'B2C'
        WHEN 'SC' THEN 'B2B'
        WHEN 'VC' THEN 'B2B'
        WHEN 'GC' THEN 'B2B'
        ELSE 'MISC Contact'
    END AS 'People Categories',
    COUNT(*) AS People
    FROM Person.Person
    WHERE PersonType NOT IN ('EM', 'SP')
    GROUP BY PersonType
) AS Customers
GROUP BY Customers.[People Categories];
"""
df_2c = pd.read_sql(query_2c_kunder, con=connection)

# Datavisualisering
# Beräkna procenttal från sql queryn 
total_people = df_2c['People'].sum()
df_2c['Percentage'] = (df_2c['People'] / total_people) * 100
plt.figure(figsize=(10, 6))
bars = plt.bar(df_2c['People Categories'], df_2c['Percentage'], color='teal')

#  Lägg till procentvärden ovanpå bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.xlabel('People Categories')
plt.ylabel('Percentage of People')
plt.title('Percentage of People by Category')
plt.xticks(rotation=45)
plt.show()
```

## Query 3: Produkter

```python
# Har tittade jag på sålda produkter, rankade dem efter både år och filtrerade som topp 10.
# Jag gjorde detta eftersom det finns två många produkter för att visualisera dem ordentligt och de mest effektiva KPIerna att 
# leta efter ledtrådar finns i de översta intervallen. 

# Jag använde en CTE (RankedProducts) för att beräkna det totala antalet sålda produkter för varje produkttyp per år och 
# rangordna dem. Sedan använde jag huvudfrågan som endast väljer de 10 bästa produkterna för varje år baserat på denna 
# rangordning. Genom att bara visualisera de 10 bästa produkterna för varje år är det lätt att se de bästa sålda produkterna 
# för att förstå AWs fokusmarknad.

# Den första rad använder en fönsterfunktion (RANK()) för att tilldela en rangordning till varje produkttyp inom varje år.
# PARTITION BY YEAR(SOH.OrderDate): Denna del av fönsterfunktionen partitionerar (eller grupperar) data efter år, 
# så att rangordningen görs separat för varje år.

query_5_produkter = """
WITH RankedProducts AS 
    (SELECT YEAR(SOH.OrderDate) AS Year
        , PS.Name AS ProductTypeName
        , SUM(SOD.OrderQty) AS TotalProductsSold
        , RANK() OVER (PARTITION BY YEAR(SOH.OrderDate) ORDER BY SUM(SOD.OrderQty) DESC) AS Rank
    FROM 
        Sales.SalesOrderHeader AS SOH
    JOIN 
        Sales.SalesOrderDetail AS SOD 
        ON SOH.SalesOrderID = SOD.SalesOrderID
    JOIN 
        Production.Product AS P 
        ON SOD.ProductID = P.ProductID
    JOIN 
        Production.ProductSubcategory AS PS 
        ON P.ProductSubcategoryID = PS.ProductSubcategoryID
    GROUP BY 
        YEAR(SOH.OrderDate), PS.Name)
SELECT Year, ProductTypeName, TotalProductsSold
FROM RankedProducts
WHERE 
    Rank <= 10
ORDER BY 
    Year, Rank;
"""

# Kör Queryn och hämta data till en DataFrame
df_5_produkter = pd.read_sql(query_5_produkter, engine)

# Pivotera DataFrame för att förbereda för plottning
pivot_df = df_5_produkter.pivot(index='Year', columns='ProductTypeName', values='TotalProductsSold')

# Skapa ett line plot
plt.figure(figsize=(12, 8))
for column in pivot_df:
    plt.plot(pivot_df.index, pivot_df[column], marker='o', linestyle='-', label=column)

plt.xlabel('Year')
plt.ylabel('Total Products Sold')
plt.title('Total Products Sold by AdventureWorks Each Year by Product Type')
plt.legend(title='Product Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(pivot_df.index, rotation=45)  # Rotera x-axelns etiketter för bättre läsbarhet
plt.grid(True)
plt.tight_layout()  
plt.show()
```

* Vi ser här de mest sålda produkterna och noterar särskilt att landsvägscyklar är den mest sålda produkten åtminstone när det gäller när företaget grundades 2011 fram till slutet av 2013. Något verkar hända med orderingången som sjönk under Q3 2013 och övergick i 2014, med cykeldelar som däck och slangar i täten.

```python
# Här fokuserar jag på de insikter jag fick från den tidigare frågan, med andra ord att landsvägscyklar är den 
# mest sålda produkten.

road_bikes_query = """
WITH SubcategorySales AS (
    SELECT YEAR(SOH.OrderDate) AS Year
            , PSC.Name AS SubcategoryName
            , SUM(SOD.OrderQty * SOD.UnitPrice) AS TotalSales
    FROM Sales.SalesOrderDetail SOD
    JOIN 
        Sales.SalesOrderHeader SOH ON SOD.SalesOrderID = SOH.SalesOrderID
    JOIN 
        Production.Product P ON SOD.ProductID = P.ProductID
    JOIN 
        Production.ProductSubcategory PSC ON P.ProductSubcategoryID = PSC.ProductSubcategoryID
    GROUP BY 
        YEAR(SOH.OrderDate), PSC.Name
)
SELECT TOP 1
    Year,
    SubcategoryName,
    TotalSales
FROM 
    SubcategorySales
ORDER BY 
    TotalSales DESC;
"""
# Kör Queryn och hämta data till en DataFrame
df_6_road_bikes = pd.read_sql(road_bikes_query, engine)
df_6_road_bikes
```

## Query 4: Försäljningsomsättning

```python
# Som den sista delen av företagsöversikten, tittade jag på försäljning och i synnerhet försäljningsomsättning som ett avgörande
# KPI-mått. Jag bestämde mig för att gräva djupare här i början och gå bortom tabellen SalesOrder Header och in 
# i SalesOrderDetail för att länka rabatterna till beräkningen av försäljningsintäkter. Det finns andra faktorer att ta hänsyn 
# till när det gäller avkastning och skattesatser, men detta ger oss en bra start.

query_4_sales_turn = """
SELECT YEAR(SOH.OrderDate) AS Year
     , SUM(SOD.OrderQty * (SOD.UnitPrice - SOD.UnitPriceDiscount)) AS SalesTurnover
FROM  Sales.SalesOrderDetail AS SOD
	JOIN Sales.SalesOrderHeader AS SOH 
	ON SOD.SalesOrderID = SOH.SalesOrderID
GROUP BY YEAR(SOH.OrderDate)
HAVING YEAR(SOH.OrderDate) BETWEEN 2011 AND 2014
ORDER BY YEAR(SOH.OrderDate);
"""
df_sales_turn = pd.read_sql(query_4_sales_turn, con=connection)
df_sales_turn
```

## Försäljnings Analys

```python
# Vi har beslutat att fokusera vår statistiska analys på försäljningsmått och det första målet är när försäljningen 
# faktiskt genomfördes för att se om det saknas data.

query_order_dates = """
SELECT MIN(OrderDate) AS FirstOrderDate
     , MAX(OrderDate) AS LastOrderDate 
FROM  Sales.SalesOrderHeader;
"""
df_odates = pd.read_sql(query_order_dates, engine)
df_odates
```

* Vi ser här att uppgifterna för 2011 börjar den 31 maj och för 2014 endast gäller för de första 6 månaderna. Detta förklarar en del av nedgångarna i försäljningen för dessa år som vi såg tidigare. Så för att göra en giltig jämförelse tittar vi istället på H1 för varje år.

```python
# Jag fokuserade på de första sex månaderna varje år för att ha jämförbara datapunkter för 2014 som den senaste datan och 
# med hänsyn till att vi redan såg något konstigt med 2014 års sålda produkter

H1_turnover_query = """
SELECT YEAR(SOH.OrderDate) AS 'Yearly H1'
     , SUM(SOD.OrderQty * (SOD.UnitPrice - SOD.UnitPriceDiscount)) AS 'Sales Turnover'
FROM  Sales.SalesOrderDetail AS SOD
	JOIN Sales.SalesOrderHeader AS SOH 
	ON SOD.SalesOrderID = SOH.SalesOrderID
WHERE MONTH(SOH.OrderDate) BETWEEN 1 AND 6
GROUP BY YEAR(SOH.OrderDate)
ORDER BY YEAR(SOH.OrderDate);
"""
df_turnover_H1 = pd.read_sql(H1_turnover_query, engine)
df_turnover_H1
```

* Vi kan bortse från 2011 eftersom aktiviteterna här endast pågick under H2, och därför ser vi en liten ökning varje år från 2012 till 2014. Det är lättare att förstå detta om vi kan representera förändringen med hjälp av jämförbara H1-ökningar från år till år.

```python
# Beräkna procentuell ökning från år till år
# Kolla att datatyperna är korrekta
df_turnover_H1['Yearly H1'] = df_turnover_H1['Yearly H1'].astype(int)
df_turnover_H1['Sales Turnover'] = df_turnover_H1['Sales Turnover'].astype(float)

# Sortera DataFrame efter År i stigande ordning
df_turnover_H1.sort_values(by='Yearly H1', inplace=True)

# Beräkna procentuell ökning år-till-år
df_turnover_H1['y_on_y_increase'] = df_turnover_H1['Sales Turnover'].pct_change() * 100
df_turnover_H1
```

* Vi ser här att det årliga tillväxtvärdet för 2011 saknas eftersom det inte fanns några tidigare försäljningsaktiviteter under 2010.

* Mellan 2011 och 2012 skedde en enorm ökning, men det beror på att 2011 års aktiviteter inleddes först den 31 maj 2011.

* De enda jämförbara uppgifterna är alltså mellan H1 2012 och H1 2013 samt mellan H1 2013 och H1 2014. Detta visar en nedgång i omsättningen på drygt 1,5%. Här måste vi alltså undersöka saken närmare. Även om det rör sig om en relativt liten minskning, så händer det något märkligt mellan 2013 och 2014.

```python
# Query för att hämta historiska försäljningsdata, så att vi kan se övergripande försäljningstrender över tid och arbeta oss 
# fram till sannolika siffror för Q3 och Q4 2014.

querySOH = """ 
SELECT SUM(TotalDue) AS TotalSales
    , YEAR(OrderDate) AS Year
    , MONTH(OrderDate) AS Month
FROM Sales.SalesOrderHeader
GROUP BY YEAR(OrderDate), MONTH(OrderDate)
ORDER BY YEAR(OrderDate), MONTH(OrderDate);
"""

df_sales = pd.read_sql(querySOH, engine)

# Skapa ett datetime-index för att säkerställa att datumstrukturer är optimerade för pythonberäkningar/visualisering 
df_sales['Date'] = pd.to_datetime(df_sales[['Year', 'Month']].assign(DAY=1))
df_sales.set_index('Date', inplace=True)

plt.figure(figsize=(12, 6))
df_sales['TotalSales'].plot(marker='', color='blue', linewidth=2, label='Sales')

plt.title('Historical Sales Data')
plt.xlabel('Date')
plt.ylabel('Total Sales (million)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Här skapar jag ett konfidensintervall med hjälp av medelvärde, standardavvikelse och observerade värden (månader) för att 
# få ett intervall för historiska försäljningssiffror. Med hjälp av detta, den tidigare försäljningslinjen och den årliga 
# minskningen på 1,5 % mellan 2013 och 2014 kan jag göra en preliminär analys för budgetering av nästa kvartal.

historical_avg_sales = df_sales['TotalSales'].mean()

std_dev = df_sales['TotalSales'].std()

n = len(df_sales)

confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

margin_error = z_critical * (std_dev / np.sqrt(n))

ci_lower = historical_avg_sales - margin_error
ci_upper = historical_avg_sales + margin_error

print(f"Variation in Sales between 2011 and 2014: {std_dev}")
print(f"Sample Size in months: {n}")
print(f"Sales based on historical average: {historical_avg_sales}")
print(f"95% Confidence Interval: ({ci_lower}, {ci_upper})")
```

* Baserat på våra historiska försäljningsdata är vi 95% säkra på att den verkliga genomsnittliga försäljningen ligger mellan ca 27 och 38 miljoner USD. Det finns en relativ normalfördelning i grafen fram till slutet av 2013 och sedan är variationen i försäljningen härifrån till slutet av Q2 2014 extremt varierad. 

* Vi ser här att försäljningen ökade dramatiskt både upp och ner under H1 2014, särskilt mellan februari och mars - över 8 miljoner USD - och sedan rasade i april och nådde en rekordlåg nivå nära 2 miljoner USD.

* Med dessa detaljer i åtanke skulle jag vilja hävda att vi bör sätta en konservativ budget för Q3 2014, med försäljningsmål på max 27 miljoner USD.

## Inventarieanalys (preliminär granskning):

```python
# Query för att hämta genomsnittliga ledtider för lager baserat på historiska data

# Här bestämde jag mig för att kontrollera ledtider för lager som ett billigt sätt för företaget att snabbt kontrollera 
# om det finns ett problem och också kunna åtgärda det snabbt utan att behöva genomgå en större kostsam och tidskrävande 
# undersökning.

# Ladda försäljningsorderdata i en DataFrame
df_sales_orders = pd.read_sql(sql="SELECT * FROM Sales.SalesOrderHeader", con=connection)

# Se till att 'OrderDate' och 'ShipDate' är i datetime-format
df_sales_orders['OrderDate'] = pd.to_datetime(df_sales_orders['OrderDate'])
df_sales_orders['ShipDate'] = pd.to_datetime(df_sales_orders['ShipDate'])

# Beräkna ledtid
df_sales_orders['LeadTime'] = (df_sales_orders['ShipDate'] - df_sales_orders['OrderDate']).dt.days

# Beräkna genomsnittlig ledtid
average_lead_time = df_sales_orders['LeadTime'].mean()

# Visa genomsnittlig ledtid
print(f"Average Lead Time: {average_lead_time} days")
```

* Här kan vi alltså se att den genomsnittliga ledtiden för försäljning av alla produkter under de 4 åren är 7 dagar. Detta verkar rimligt men vi måste kontrollera om det finns några variationer över tiden för att avgöra om justeringar behövs eller inte.

```python
# Sortera DataFrame efter OrderDate för ett kronologiskt diagram så att vi kan kontrollera om det finns några problem med 
# ledtider som påverkar försäljningen.

df_sales_orders = df_sales_orders.sort_values(by='OrderDate')

# Create a line plot
plt.figure(figsize=(12, 7))
plt.plot(df_sales_orders['OrderDate'], df_sales_orders['LeadTime'], marker='o', color='skyblue', linestyle='-', markersize=3)
plt.title('Lead Times Over Time')
plt.xlabel('Order Date')
plt.ylabel('Lead Time (days)')
plt.grid(True)
plt.show()
```

* Ledtiderna här verkar vara ganska stabila från 2011 till slutet av 2013. En liten variation sker under 2014 från 7 dagar till 8 dagar. Jag rekommenderar att du kontrollerar om det finns en märkbar skillnad i försäljningsvärden mellan kategorierna i de fyra produktlinjerna för att avgöra om det skulle vara bättre att fokusera på några utvalda toppsäljande produkter.

```python
# Här har jag genomfört ett hypotestest för att kontrollera om mellan de fyra produktlinjerna (vi tar två som oberoende 
# stickprov) har en betydande skillnad i försäljningspris som skulle innebära att det skulle vara mer meningsfullt att titta 
# på specifika och avgörande produkter som har en proportionellt större inverkan på ledtiderna.

# Nollhypotes (H0): Det finns ingen signifikant skillnad i genomsnittligt försäljningsbelopp mellan produktkategorier .
# Alternativ hypotes (H1): Det finns en signifikant skillnad i genomsnittligt försäljningsbelopp mellan produktkategorier.

query4 = """
SELECT 
    PC.Name AS CategoryName,
    SOD.UnitPrice * SOD.OrderQty AS SaleAmount
FROM 
    Sales.SalesOrderDetail SOD
JOIN 
    Production.Product P ON SOD.ProductID = P.ProductID
JOIN 
    Production.ProductSubcategory PSC ON P.ProductSubcategoryID = PSC.ProductSubcategoryID
JOIN 
    Production.ProductCategory PC ON PSC.ProductCategoryID = PC.ProductCategoryID
WHERE 
    PC.Name IN ('Bikes', 'Accessories');
"""

df4 = pd.read_sql(query4, engine)

avg_sales = df4.groupby('CategoryName')['SaleAmount'].mean()

plt.figure(figsize=(8, 6))
avg_sales.plot(kind='bar', color=['blue', 'green'])
plt.title('Average Sale Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Sale Amount')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter data for each category
sales_bikes = df4[df4['CategoryName'] == 'Bikes']['SaleAmount']
sales_accessories = df4[df4['CategoryName'] == 'Accessories']['SaleAmount']

# Perform an independent t-test
t_stat, p_val = stats.ttest_ind(sales_bikes, sales_accessories, equal_var=False)  # Welch’s t-test

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

- P-värdet är 0 så vi kan förkasta nollhypotesen och vi konstaterar att det finns en stor skillnad i försäljning mellan kategorierna
- En positiv T-statistik antyder att den första gruppen har ett högre medelvärde, så vi kommer att fokusera på ledtider för underkategorier under denna kategori, nämligen "bikes". 

```python
# I den här koden använder jag python för att beräkna den genomsnittliga ledtiden och kontrollera den mot ett antal 
# dataframeserier för att gruppera de 5 viktigaste underkategorierna inom "Bikes". 
# Jag gör detta så att vi kan gräva djupare och se vad som verkar orsaka problemet med ledtider.

leads_bikes_query = """
SELECT soh.OrderDate
	, soh.ShipDate
	, psc.Name AS SubCategoryName
	, sod.OrderQty
	, sod.UnitPrice
	, (sod.OrderQty * sod.UnitPrice) AS TotalSales
FROM Sales.SalesOrderHeader AS soh
JOIN Sales.SalesOrderDetail AS sod 
ON soh.SalesOrderID = sod.SalesOrderID
JOIN Production.Product AS p 
ON sod.ProductID = p.ProductID
JOIN Production.ProductSubcategory AS psc 
ON p.ProductSubcategoryID = psc.ProductSubCategoryID;
"""

df_sales_leads = pd.read_sql(sql=leads_bikes_query, con=connection)

# Konvertera 'OrderDate' och 'ShipDate' till datetime-format
df_sales_leads['OrderDate'] = pd.to_datetime(df_sales_leads['OrderDate'])
df_sales_leads['ShipDate'] = pd.to_datetime(df_sales_leads['ShipDate'])

# Beräkna ledtid
df_sales_leads['LeadTime'] = (df_sales_leads['ShipDate'] - df_sales_leads['OrderDate']).dt.days

# Beräkna totalförsäljning för varje underkategori
subcategory_sales = df_sales_leads.groupby('SubCategoryName')['TotalSales'].sum().reset_index()

# Identifiera de mest säljande underkategorierna (exempel: topp 5 baserat på total försäljning)
top_subcategories = subcategory_sales.nlargest(5, 'TotalSales')['SubCategoryName']

# Filtrera den ursprungliga dataframe:n för att endast inkludera de mest säljande underkategorierna
top_sales_orders = df_sales_leads[df_sales_leads['SubCategoryName'].isin(top_subcategories)]

# Gruppera efter 'OrderDate' och 'SubCategoryName' för de mest säljande underkategorierna
grouped_data = top_sales_orders.groupby(['OrderDate', 'SubCategoryName'])['LeadTime'].mean().reset_index()

plt.figure(figsize=(15, 7))
for category in grouped_data['SubCategoryName'].unique():
    category_data = grouped_data[grouped_data['SubCategoryName'] == category]
    plt.plot(category_data['OrderDate'], category_data['LeadTime'], marker='o', label=category)
plt.title('Lead Times Over Time by Top Selling Product Subcategories (Based on Total Sales)')
plt.xlabel('Order Date')
plt.ylabel('Average Lead Time (days)')
plt.legend()
plt.grid(False)
plt.show()
```

- Vi kan nu se att landsvägsramar och landsvägscyklar har störst påverkan på ledtiderna, även om skillnaden är ganska liten, dvs. mellan 0,2 och 1 dag över genomsnittet. 
- Denna query tillsammans med det faktum att vi vet att försäljningen av landsvägscyklar har minskat kraftigt sedan mitten av 2013 innebär dock att det krävs en djupare undersökning mellan avdelningarna för att ta reda på om detta beror på interna problem inom produktion, försäljning och marknadsföring eller externa marknadsvillkor, eller en blandning av båda.

```python
# Denna query denna fråga att ge en sammanfattad vy av den totala försäljningen uppdelad efter produktkategori och kvartal, 
# vilket är användbart för att analysera försäljningstrender över tid för olika produktlinjer.

quarter_sales_query = """
SELECT psc.Name AS CategoryName
    , CONCAT(YEAR(soh.OrderDate), ' Q', CEILING(MONTH(soh.OrderDate) / 3.0)) AS Quarter
    , SUM(sod.OrderQty * sod.UnitPrice) AS TotalSales
FROM  Sales.SalesOrderHeader AS soh
    JOIN Sales.SalesOrderDetail AS sod ON 
    soh.SalesOrderID = sod.SalesOrderID
    JOIN 
    Production.Product AS p 
    ON sod.ProductID = p.ProductID
    JOIN 
    Production.ProductSubcategory AS psc 
    ON p.ProductSubcategoryID = psc.ProductSubCategoryID
GROUP BY 
    psc.Name, YEAR(soh.OrderDate), CEILING(MONTH(soh.OrderDate) / 3.0)
ORDER BY 
    psc.Name, YEAR(soh.OrderDate), CEILING(MONTH(soh.OrderDate) / 3.0);

"""

df_category_sales = pd.read_sql(sql=quarter_sales_query, con=engine)

# Filtrera för de 3 största kategorierna baserat på total försäljning
top_categories = df_category_sales.groupby('CategoryName')['TotalSales'].sum().nlargest(3).index

# Filtrera DataFrame för att få de 3 bästa kategorierna
df_top_categories = df_category_sales[df_category_sales['CategoryName'].isin(top_categories)]

plt.figure(figsize=(15, 7))
for category in top_categories:
    category_data = df_top_categories[df_top_categories['CategoryName'] == category]
    plt.plot(category_data['Quarter'], category_data['TotalSales'], marker='o', label=category)
plt.title('Sales Trends for Top 3 Categories')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

* Om vi tittar närmare på kvartal över tid är de försäljningstrender vi kan utläsa från dessa bäst säljande produkter att sommarsäsongen mellan Q2 och Q3 verkar uppvisa säsongsmässiga uppgångar mellan 2011 och 2013. Så trots den nedåtgående försäljningsspiralen under H1 2014, kan detta lätta på trycket när företaget går in i sommarsäsongen.
